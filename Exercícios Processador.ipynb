Uma UCP é composta por partes como o registrador de instruções(RI), unidade lógica aritmética, REM, RDM e registradores de dados. Uma UCP pode fazer operações aritméticas como somar, subtrair, multiplicar e dividir, pode manipular dados mudando eles de lugar na memória e também transferir informações para um dispositivo de saída.

A função da unidade aritmética e lógica é lidar com operações primitivas como adição, subtração e também operações lógicas com AND ou OR.

A uniade de controle de um processador é respondável por medir basicamente  todas as ações de um processador. Ela define se haverá uma operação de leitura ou escrita na MP e quais movimentações internas e externas executar.

CI ou PC. Nele se armazena o endereço da próxima instrução a ser executada.

Para o ADD Op. temos dois ciclos de memória, sendo o primeiro para ler a instrução e a segunda para ler o operando e completar a operação (dois ciclos de leitura).
No SUB Op. temos três ciclos de memória. Um para ler a instrução, outro para ler o operando e o terceiro para gravar o resultado da operação no endereço original do operando (dois ciclos de leitura e um de escrita).
No ADD Op.1, Op.2 temos 4 ciclos de memória. Um para buscar a instrução, um para ler o primeiro operando, o terceiro para ler o segundo operando e o quarto para escrever o resultado no endereço original do operando 1 (três ciclos de leitura e um de escrita).
No INCR temos apenas um ciclo de memória, que é para ler a instrução.
No LDA Op. temos dois ciclos de memória. Sendo um para ler a instrução e o outro para ler o operando no endereço especificado.
